## List - one variable with many values

`animals = ["cat", "dog", "fish"]`

A list can hold many values at once - the values can be any type and the list can be any length.  It is held inside `[]` and each value is separated by a comma `,`.

The values in a list are numbered, starting with `0`.

See the example below:

````python
animals = ["cat", "dog", "fish"]
print(animals[0])
print(animals[1])
print(animals[2])
print(len(animals)) # how many animals are in the list?
animals.append("cow")
print(animals)
````

## Loops

Suppose you want your computer to do the same thing many times.  You use a **loop**.  There are many styles of loops.  This one is called a `for` loop.


````python
animals = ["cow", "horse", "goat"]
for animal in animals:
    print(animal)

for i in range(len(animals)):
    print(animals[i])
````

## Decisions

Sometimes you need the computer to do something different depending on certain *conditions*.  For example, in a computer game:

```
if score == 0:
    print("You lose")
elif score < 100:
    print("Stay on level 1")
else:
    print("Go to the next level")
```

Notice the `:` at the end of each line, and the **indentation**, which tells Python what to do when the *condition* is true.

Condition or **relational operators** are `<`, `<`, `!=`, `>=`, `<=`, and `==`.  The result is either True or False.

Notice how in Python, `==` means equals, while `=` means put a value in a variable.

````python
# This is a simple animal classifier

legs = 8

if legs == 8:
    print("It might be a spider! 🕷️")
elif legs == 4:
    print("Maybe a cat, dog, or horse?")
elif legs == 2:
    print("A bird... or a human!")
else:
    print("Hmm, a snake? A fish?")
````

